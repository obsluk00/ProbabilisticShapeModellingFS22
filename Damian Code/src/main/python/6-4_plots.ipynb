```python
import matplotlib.pyplot as plt
import pandas as pd

dataDf = pd.read_csv("/home/damian/Documents/Uni/Probabilistic Shape Modelling/tutorial-14/results/classifiedSamples.csv")
femaleDf = dataDf[dataDf[' sex'] == ' f']
maleDf = dataDf[dataDf[' sex'] == ' m']

trainingDf = pd.read_csv("/home/damian/Documents/Uni/Probabilistic Shape Modelling/tutorial-14/Resources/statureAndSex.csv")
trainingDf = trainingDf.rename(columns={" bone-length":" length", " trochanter-distance": " width"})

trainingFemaleDf = trainingDf[trainingDf[' sex'] == ' f']
trainingMaleDf = trainingDf[trainingDf[' sex'] == ' m']
```

```python
femaleAxis = femaleDf.plot(kind='scatter', x=' length', y=' width', color='Pink', label="female")
maleAxis = maleDf.plot(ax=femaleAxis, kind='scatter', x=' length', y=' width', color='Blue', label="male")
trainingFemaleAxis = trainingFemaleDf.plot(ax=maleAxis, kind='scatter', x=' length', y=' width', color='Green', label=' training-female')
trainingMaleAxis = trainingMaleDf.plot(ax=trainingFemaleAxis, kind='scatter', x=' length', y=' width', color='Orange', label=' training-male')
```
